```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

# Load the cleaned data
df = pd.read_csv("Clinical Dataset_PreP_Output.csv")

# 1. Map Sex to Numeric (Male: 0, Female: 1, All: 0.5)
sex_map = {'MALE': 0, 'FEMALE': 1, 'ALL': 0.5}
df['Sex_Numeric'] = df['Sex'].map(sex_map)

# 2. Map Results to Binary (Target)
df['Result_Binary'] = df['Study Results'].apply(lambda x: 1 if x == 'YES' else 0)

# 3. Handle Phases - Simplified mapping for numeric weight
phase_map = {'PHASE1': 1, 'PHASE2': 2, 'PHASE3': 3, 'PHASE4': 4, 'EARLY_PHASE1': 0.5}
df['Phase_Numeric'] = df['Phases'].map(phase_map).fillna(0)

# 4. Remove the NAs we discussed
df_ml = df[df['Phase_Numeric'] > 0].copy()

# 5. Fit the Model
X = df_ml[['Phase_Numeric', 'Sex_Numeric', 'Enrollment']]
y = df_ml['Result_Binary']
model = LogisticRegression().fit(X, y)

print(f"Intercept: {model.intercept_[0]}")
print(f"Coefficients (Phase, Sex, Enrollment): {model.coef_[0]}")
```
